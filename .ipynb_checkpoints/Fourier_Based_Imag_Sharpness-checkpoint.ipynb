```python
import numpy as np
import cv2
import PIL
from PIL import Image
import matplotlib.pyplot as plt
import os
import glob
```

```python
imgs = glob.glob("H:/Python/Research Practise/Notebooks/image_sharpness_frequency_domain/Test_Data/test7/"+"*.jpg")
```

```python
def focus(img):
    img = cv2.imread(img, cv2.IMREAD_GRAYSCALE)
    print(img.shape)
    f = np.fft.fft2(img)
    fshift = np.fft.fftshift(f)
    print(fshift.shape)
    magnitude_spectrum = 20 * np.log(np.abs(fshift))
    
    x = np.linspace(0, 1, 325)
    y = np.linspace(0, 1, 217)
    X, Y = np.meshgrid(x, y)
    
    highpass = 1 - np.exp(- ((X - 0.5) ** 2 + (Y - 0.5) ** 2) * 5)
    print(np.shape(highpass))
    f2 = fshift * highpass
    z3 = np.absolute(np.fft.ifft2(f2))
    
    plt.subplot(337)
    plt.imshow(z3)
    plt.title('only high frequency content survived')
    plt.colorbar()
    plt.subplot(338)
    plt.imshow(highpass)
    plt.title('highpass, suppresses \n low frequencies')
    plt.colorbar()
    plt.subplot(339)
    plt.imshow(np.log10(np.abs(fshift * highpass)), cmap='gray')
    plt.title('FFT*highpass')
    plt.colorbar()
    plt.show()
```

```python
img = cv2.imread(imgs[0], cv2.IMREAD_GRAYSCALE)
print(img.shape)
f = np.fft.fft2(img)
fshift = np.fft.fftshift(f)
print(fshift.shape)
magnitude_spectrum = 20 * np.log(np.abs(fshift))

x = np.linspace(0, 1, 325)
y = np.linspace(0, 1, 217)
X, Y = np.meshgrid(x, y)

highpass = 1 - np.exp(- ((X - 0.5) ** 2 + (Y - 0.5) ** 2) * 5)
print(np.shape(highpass))
f2 = fshift * highpass
z3 = np.absolute(np.fft.ifft2(f2))

fig, ax = plt.subplots(3, figsize=(10,10))
fig.subplots_adjust(wspace = 0.4, hspace = 0.4)

ax[0].imshow(z3)
ax[0].set_title('only high frequency content survived')


ax[1].imshow(highpass, )
ax[1].set_title('highpass filter, suppresses \n low frequencies')
# ax[1].colorbar()

ax[2].imshow(np.log10(np.abs(fshift * highpass)), cmap='gray')
ax[2].set_title('FFT*highpass')
# ax[2].colorbar()
```

```python
vals = []
for i in range(len(imgs)):
    img = cv2.imread(imgs[i], cv2.IMREAD_GRAYSCALE)
    f = np.fft.fft2(img)
    fshift = np.fft.fftshift(f)
    magnitude_spectrum = 20 * np.log(np.abs(fshift))
    
    x = np.linspace(0, 1, 325)
    y = np.linspace(0, 1, 217)
    X, Y = np.meshgrid(x, y)
    
    highpass = 1 - np.exp(- ((X - 0.5) ** 2 + (Y - 0.5) ** 2) * 5)
    # print(np.shape(highpass))
    f2 = fshift * highpass
    z3 = np.absolute(np.fft.ifft2(f2))
    vals.append(np.sum(z3))
```

```python
plt.plot(vals)
plt.title("Variation of magnitude of high frequencies remaining \n after applying high pass filter") 
plt.ylabel("Sum of magnitude of all pixels")
plt.xlabel("Image index")
```

```python
print("Image index with highest value", np.argmax(vals))
```
